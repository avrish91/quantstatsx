```python
%matplotlib inline
import pandas as pd
import quantstats as qs
```

```python
# Grab In Sample Spreader SIMs
df1 = pd.read_parquet('C:/Users/User/quantstatsx/is_sim_variable_gmv_35_pov_nof_part1.pqt')
df2 = pd.read_parquet('C:/Users/User/quantstatsx/is_sim_variable_gmv_35_pov_nof_part2.pqt')
df3 = pd.read_parquet('C:/Users/User/quantstatsx/is_sim_variable_gmv_35_pov_nof_part3.pqt')
df4 = pd.read_parquet('C:/Users/User/quantstatsx/is_sim_variable_gmv_35_pov_nof_part4.pqt')
is_df = pd.concat([df1,df2,df3,df4])
oos_df = pd.read_parquet('C:/Users/User/quantstatsx/oos_sim_variable_gmv_35_pov_nof.pqt')
```

```python
capital = 1e6
is_df['agg_pnl'] = is_df['cum_net_pnl']
# Combine pnl
oos_df['agg_pnl'] = oos_df['cum_net_pnl'] + is_df['cum_net_pnl'].iloc[-1]
agg_df = pd.concat([is_df, oos_df])
agg_df['agg_wallet_bal'] = agg_df['agg_pnl'] + capital
# agg_df['agg_wallet_bal'].plot()
```

```python
agg_df_resampled = agg_df.resample('D').last()
agg_df_resampled['daily_return'] = agg_df_resampled['agg_wallet_bal'].pct_change()
# agg_df_resampled.dropna(inplace = True) !!! Careful with that
```

```python
ret_df = agg_df_resampled[['agg_wallet_bal', 'daily_return']]
ret_df.dropna(inplace = True)
```

```python
ret_df.index.names = ['Date']
```

```python
# (benchmark can be a pandas Series or ticker)
# fetch the daily returns for a stock
# stock = qs.utils.download_returns('PLTR')
qs.reports.full(ret_df['daily_return'], compounded = False)
```
